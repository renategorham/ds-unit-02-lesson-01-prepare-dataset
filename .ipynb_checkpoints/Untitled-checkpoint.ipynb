Purpose: Analyze variables to choose the most important features to develop a model to predict abalone age and sex using physical measurements.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
```

```python
abalone = pd.read_csv('abalone.csv')
abalone.drop('Unnamed: 0', axis=1, inplace=True)
abalone.head()
```

```python
abalone.shape
```

```python
abalone.dtypes
```

#### Dependent variables
The goal of this analysis is to determine the age and sex of an abalone by measuring physical characteristics.  
The dataset has 4177 measures. Of those, 1307 were determined to be female, 1528 male and 1342 were indeterminate.

```python
sex_count = abalone.groupby('Type').count()
sex_count
```

Abalones and other gastropods lay down a layer of calcium carbonate each year. Counting the rings of a shell works the same as counting the rings on a tree. One ring equals one year. The median age is 9 years, the minimum and maximum are 1 and 29 years.

```python
abalone.Rings.describe()
```

```python
plt.hist(abalone.Rings)
plt.title('Histogram of Abalone Age')
plt.xlabel('Number of Rings / Age')
plt.ylabel('Count')
plt.show()
```

#### Univariate analysis

```python
abalone_factors = abalone.loc[:,'LongestShell':'ShellWeight']
```

The chart below shows pair-wise comparisons of independent variables. As my be expect, the first three variables, height, longest shell length and shell diameter appear to be highly linearly correlated. The same trend can be observed in the weight variables.  

Initially, choose the least invasive measurements. In this case, to develop a model to determine sex and age of an abalone, choose longest shell length and whole weight.

```python
sns.set_style('whitegrid')

g = sns.PairGrid(abalone_factors.dropna(), diag_sharey=False)
g.map_upper(plt.scatter, alpha=0.5)
g.map_lower(sns.regplot, scatter_kws=dict(alpha=0))
g.map_diag(sns.kdeplot, lw=3)
plt.subplots_adjust(top=0.95)
g.fig.suptitle('Matrix Plot of Pair-wise Factors', fontsize='20')
plt.show()
```

#### Multivariate analysis

Comparisons between the measurements for the dependent variable categories are shown below. 

```python
abalone_by_sex = abalone.loc[:,'Type':'ShellWeight']
abalone_by_sex = pd.melt(abalone_by_sex, id_vars=['Type'])
g = sns.FacetGrid(abalone_by_sex, col='variable', height=4, aspect=0.5)
g = g.map(sns.boxplot, 'Type', 'value', order=None)
plt.show()
```

Analysis of variance indicates that there is a statistical difference between at least one of the categories of the dependent variables. The boxplot above seems to indicate that indeterminate has the lower values.

```python
for col in abalone.loc[:,'LongestShell':'ShellWeight'].columns:
    print(col)
    print(stats.f_oneway(
        abalone[abalone['Type']=='M'][col],
        abalone[abalone['Type']=='F'][col],
        abalone[abalone['Type']=='I'][col]
    ))
```

A t-test shows that there is also a significant difference between male and female types.

```python
for col in abalone.loc[:,'LongestShell':'ShellWeight'].columns:
    print(col)
    print(stats.ttest_ind(
        abalone[abalone['Type']=='M'][col],
        abalone[abalone['Type']=='F'][col]
    ))   
```

#### Relationship between dependent variables

```python
plt.hist(abalone[abalone['Type']=='M']['Rings'])
plt.hist(abalone[abalone['Type']=='F']['Rings'])
plt.hist(abalone[abalone['Type']=='I']['Rings'])
plt.title('Histogram of Rings by Type')
plt.xlabel('Number of Rings')
plt.ylabel('Count')
plt.show()
        
```

Comparisons of the number of rings (age) by sex, shows that at least one of the types or age groups is different.

```python
stats.kruskal(abalone[abalone['Type']=='M']['Rings'],\
              abalone[abalone['Type']=='F']['Rings'],\
             abalone[abalone['Type']=='I']['Rings'])
```

The Krustkal-Wallis test only says that at least one of the groups are different. The Mann-Whitney test below inicates that there is a difference between males and females.

```python
stats.mannwhitneyu(abalone[abalone['Type']=='M']['Rings'],\
                   abalone[abalone['Type']=='F']['Rings'])
```

#### Conclusion
The main features of the model may be longest shell length and total weight to predict the sex and age of abalone.
